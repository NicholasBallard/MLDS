# Plotting

In this notebook we will briefly discuss the most popular tools in Python for data visualisation, namely matplotlib and Seaborn.

## Matplotlib

First we will import matplotlib and use it to plot some dummy data. Later we will import Seaborn and see how it helps to make your plots more visually appealing. 

```python
import matplotlib.pyplot as plt # Convention
import numpy as np
import pandas as pd
# This line ensures that matplotlib plots 'inline' in the notebook (and not in a new window)
%matplotlib inline 
```

Now we generate some dummy data and see how you can make 2D plots of data quickly with matplotlib:

```python
x = np.random.randint(100, size=25)
y = x*x
```

We now have some dummy data, which we will plot using the `scatter` function: 

```python
plt.scatter(x,y);
```

## Seaborn

The appearance of plots can be improved dramtically using Seaborn. Often, all that is required is that you import Seaborn and plot as you normally would. First we import Seaborn:

```python
import seaborn as sns
plt.scatter(x,y);
```

As you can see, we have created this plot using the same function as the previous example. The results are, subjectively, better. 

Seaborn includes its own plotting tools of course. For example, you can use the `lmplot` function to plot a line that should fit to your data.

We will convert our data into a Pandas DataFrame object, the expected input of the `lmplot` function:

```python
df = pd.DataFrame({"x": x, "y": y})
df.head()
```

Now we can plot the data and fit a line to it:

```python
sns.lmplot(x="x", y="y", data=df);
```

As can be seen, the data is plotted and a linear model is made to fit the data.
